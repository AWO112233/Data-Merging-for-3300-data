```python
import pandas as pd
```

**Merging Team+Service+BusinessDeliveryUnit = Team_merge**

```python
#Load Data
team_df = pd.read_csv('Team.csv')
Service_df = pd.read_csv('Service.csv')
BDU_df = pd.read_csv('BusinessDeliveryUnit.csv')
```

```python
#Service and BDU merge
buser = Service_df.merge(BDU_df, how="left", left_on="Service_BusinessDeliveryUnitId", right_on="BusinessDeliveryUnit_Id")
```

```python
#Final Team Merge using Team+buser
Team_merge = team_df.merge(buser, how="left", left_on="Team_ServiceId", right_on="Service_Id")
```

```python
#Team Merge data save
Team_merge.to_csv('Team_merge.csv', index=False)
```

**Merging PersonBooking+Person+PersonBookingType = Person_merge**


```python
#Load data
PerB_df = pd.read_csv('PersonBooking.csv')
Per_df = pd.read_csv('Person.csv')
PerBType_df = pd.read_csv('PersonBookingType.csv')
```

```python
#PersoonBooing and Person Merge
per1 = PerB_df.merge(Per_df, how="left", left_on="PersonBooking_PersonId", right_on="Person_Id")
```

```python
#Final Merge using per1 and PerBType
Person_merge = per1.merge(PerBType_df, how="left", left_on="PersonBooking_PersonBookingTypeId", right_on="PersonBookingType_Id")
```

```python
#Person Merge data save
#Person_merge.to_csv('Person_merge.csv', index=False)
```

**Merging BookingOccasion+CancellationReason+BookingOccasionState = BookingOcc_merge**

```python
#Load Data
BookOcc = pd.read_csv('BookingOccasion.csv')
CanReas = pd.read_csv('CancellationReason.csv')
BookOccState = pd.read_csv('BookingOccasionState.csv')
```

```python
#BookOcc + CanReas
book1 = BookOcc.merge(CanReas, how="left", left_on="BookingOccasion_CancellationReasonId", right_on="CancellationReason_Id")
```

```python
#Final Merge using book1 and BookOccState
BookingOcc_merge = book1.merge(BookOccState, how="left", left_on="BookingOccasion_BookingOccasionStateId", right_on="BookingOccasionState_Id")
```

```python
# Booking Occ Result Data Save
#BookingOcc_merge.to_csv('BookingOcc_merge.csv', index=False)
```

**Area+Buiding+District+BuildingOpeningHours Merge = Area_Merge**

```python
#Load Data
Area_df = pd.read_csv('Area.csv')
Building_df = pd.read_csv('Building.csv')
District_df = pd.read_csv('District.csv')
BuildingOpeningHours_df = pd.read_csv('BuildingOpeningHours.csv')
```

```python
#merging Building_df with District.
BuilDes = Building_df.merge(District_df, how="left", left_on="Building_DistrictId", right_on="District_Id")
```

```python
#Merging BuildingOpeningHours with BuilDes
BuilOpHr_BuilDes = BuildingOpeningHours_df.merge(BuilDes, how="left", left_on="BuildingOpeningHours_BuildingId", right_on="Building_Id")
```

```python
#Merging Area_df with BuildOpHr
Area_Merge = Area_df.merge(BuilOpHr_BuilDes, how="left", left_on="Area_BuildingId", right_on="Building_Id")
```

```python
#Area merge result data save
#Area_Merge.to_csv('Area_merge.csv', index=False)
```

**BookUnit+BookVir+BookUntTyp+BookUnitEqp+Eqp+Area_Merge = BookUnit_merge**

```python
BookUnit_df = pd.read_csv('BookableUnit.csv')
BookVir_df = pd.read_csv('BookableUnitVirtual.csv')
BookUntTyp_df = pd.read_csv('BookableUnitType.csv')
BookUnitEqp_df = pd.read_csv('BookableUnitEquipment.csv')
Eqp_df = pd.read_csv('Equipment.csv')
```

```python
#BU_BUT + BookVir_df
BU_BUT_BV = BU_BUT.merge(BookVir_df, how="left", left_on="BookableUnit_Id", right_on="BookableUnitVirtual_BookableUnitId")
```

**Booking+BookType+BookState+BookAutSta+AvailSearch = booking_merge**


```python
booking_df = pd.read_csv('Booking.csv')
BookType_df = pd.read_csv('BookingType.csv')
BookState_df = pd.read_csv('BookingState.csv')
BookAutSta_df = pd.read_csv('BookingAuthorisationStatus.csv')
AvailSer_df = pd.read_csv('AvailabilitySearch.csv')
```
